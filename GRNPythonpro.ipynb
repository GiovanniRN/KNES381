```python

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
## This is the VO2 chart with the new data

df3 = pd.read_csv('../input/demo-knes381/subject_1321.csv', header=[0], skiprows=[1,2,3])

df3 = df3.rename(columns={'VE/': 'VE/VO2','VE/.1': 'VE/VCO2'})


x = df3['TIME']
y = df3['VO2']
y1 = df3['VE']

ymax = max(y)

xmax = x[y.argmax()]

fig, ax = plt.subplots(2, 1, sharex=True, figsize=(8, 10)) 


fig.subplots_adjust(hspace=0)



ax[0].annotate('$\dot VO_2max$ =({}) L/min'.format(round(ymax, 2)), 
               xy=(xmax, ymax), xytext=(xmax+.5, ymax+ .5),
               arrowprops=dict(facecolor='red', shrink= 0.05),
                )

ax[0].plot(x, y, 'o', label=('$\dot VO_2$'), c='r' )

ax[0].spines[['right', 'top']].set_visible(False)
ax[0].set(ylabel=('L/min'))
ax[0].legend()


ax[1].plot(x, y1, 'o', label=('VE'), c='b')
ax[1].spines[['top', 'right']].set_visible(False)
ax[1].set(ylabel=('breaths/min'))
ax[1].set(xlabel=('Time(min)'))
ax[1].legend()


fig.savefig("VO2-VE-GRN.png", dpi=300, bbox_inches = "tight")
fig.show()

##This code is for the FeCO2 and VCO2 graphs

df3 = pd.read_csv('../input/demo-knes381/subject_1321.csv', header=[0], skiprows=[1,2,3])

df3 = df3.rename(columns={'VE/': 'VE/VO2','VE/.1': 'VE/VCO2'})


x = df3['VO2']
y = df3['FECO2']
y1 = df3['VCO2']

ymax = max(y)

xmax = x[y.argmax()]

fig, ax = plt.subplots(2, 1, sharex=True, figsize=(8, 10)) 


fig.subplots_adjust(hspace=0.1)


ax[0].plot(x, y, 'o', label=('$\dot FECO2$'), c='r' )

ax[0].spines[['right', 'top']].set_visible(False)
ax[0].set(ylabel=('%'))
ax[0].axvline(x=2.26, color='k', linestyle='--', label='RCP')
ax[0].axvline(x=1.25, color='r', linestyle='--', label='GET')
ax[0].legend()


ax[1].plot(x, y1, 'o', label=('VCO2'), c='b')
ax[1].spines[['top', 'right']].set_visible(False)
ax[1].set(ylabel=('L/min'))
ax[1].set(xlabel=('VO2(L/min)'))
ax[1].axvline(x=2.26, color='k', linestyle='--', label='RCP')
ax[1].axvline(x=1.25, color='r', linestyle='--', label='GET')
ax[1].legend()

fig.savefig("VO2-VE-GRN2.png", dpi=300, bbox_inches = "tight")
fig.show()
```
